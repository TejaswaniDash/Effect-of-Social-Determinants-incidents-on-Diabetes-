```python
import pandas as pd
```

```python
common_data = pd.read_csv('commonSymptoms.csv' ,index_col = 0)
```

```python
common_data.head()
```

```python
import numpy as np
c_matrix = np.zeros(shape=(29,29))
print(c_matrix)
```

```python
common_data_numpy = common_data.to_numpy()
common_data_numpy = np.nan_to_num(common_data_numpy)

print(common_data_numpy)
```

```python
for i in range(0,29):
  for j in range(0,29):
    if(i == j):
      c_matrix[i,j] = 2 + np.sum(common_data_numpy[i])
    else:
      c_matrix[i,j] = -common_data_numpy[i,j]
```

```python
occurs_later_array = np.zeros(shape= (29))
occurs_before_array = np.zeros(shape = (29))
```

```python
difference_data = pd.read_csv('DifferenceSymptoms.csv',index_col = 0)
difference_data.head()
```

```python
difference_data_numpy = difference_data.to_numpy()
print(difference_data_numpy[0][0])
```

```python
difference_matrix = np.zeros(shape= (29,1))
print(difference_matrix)
```

```python
for i in range(0,29):
  w_count = 0
  l_count = 0
  for j in range(0,29):
    if(i == j):
      w_count = w_count + 0
      l_count = l_count + 0
    elif(difference_data_numpy[i][j] < difference_data_numpy[j][i]):
      w_count = w_count + 1
    elif((difference_data_numpy[i][j]) > difference_data_numpy[j][i]):
      l_count = l_count + 1
    elif((difference_data_numpy[i][j]) == difference_data_numpy[j][i]):
      l_count = l_count + 1
      w_count = w_count + 1
  res = 1 + (w_count - l_count)/2
  print(res)
  difference_matrix[i] = res 
```

```python
print(difference_matrix.shape)
print(c_matrix.shape)
```

```python
c_matrix_inv = np.linalg.inv(c_matrix)
print(c_matrix_inv.shape)
```

```python
rank_matrix = np.dot(a=c_matrix_inv,b=difference_matrix)
print(rank_matrix)
```

```python
my_list = ['LossSmell', 'Fatigue', 'Difficultybreathing',
       'SlurredSpeech', 'Diarrhea', 'LossAppetite', 'Bluish', 'MuscleAches',
       'Confusion', 'Pinkeye', 'UnexplainedRash', 'Shortnessofbreath',
       'JointPain', 'Sorethroat', 'Shivering', 'AbdominalPain', 'ExcessSweat',
       'LossBalance', 'RedRash', 'Numbness', 'Headaches', 'Cough', 'Chills',
       'LossTaste', 'Wheezing', 'Fever', 'Vomiting', 'Runnynose', 'Chestpain']
```

```python
print(len(my_list))
my_rank_data = pd.DataFrame(data=rank_matrix,columns=['Rank Score'],index=my_list)

my_rank_data.head()
my_rank_data.to_csv('Rank_Score_Data.csv')
```

```python
my_rank_data_sorted = my_rank_data.sort_values(by=['Rank Score'],ascending=False)
my_rank_data_sorted.head()

my_rank_data_sorted.to_csv('Rank_Score_Data_Sorted_Asc.csv')
```

```python
rank_ordering_list = (my_rank_data_sorted.index)
print(rank_ordering_list)
```
